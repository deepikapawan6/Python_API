```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

column_names = ["City", "Country"]
cities_df = pd.DataFrame(columns=column_names)
```

```python
for x in range(1500):
    lat = np.random.uniform(low=-90.000, high=90.000) 
    lng = np.random.uniform(low=-180.000, high=180.000)
    city = citipy.nearest_city(lat, lng)
    cities_df = cities_df.append({"City": city.city_name,"Country": city.country_code,}, ignore_index=True)

#Remove duplicate cities
cities_df = cities_df.drop_duplicates(subset='City').reset_index()
cities_df.count()
```

```python
#Create a df to add all the data from the API calls.
column_names = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]
citypy_df = pd.DataFrame(columns=column_names)

# Loop through the list of cities and add the data to the df.
url = "http://api.openweathermap.org/data/2.5/weather?"

row_count = 1

print("Beginning Data Retrieval")
print("------------------------")
for index, row in cities_df.iterrows():
    target_city = row["City"]
    target_url = f"{url}appid={api_key}&units=IMPERIAL&q={target_city}"
    
    response = requests.get(target_url).json()
    
    if response["cod"] == "404":
        print("City not found. Skipping...")

    else:
        citypy_df.set_value(index, "City", response["name"])
        citypy_df.set_value(index, "Cloudiness", response["clouds"]["all"])
        citypy_df.set_value(index, "Country", response["sys"]["country"])
        citypy_df.set_value(index, "Date", response["dt"])
        citypy_df.set_value(index, "Humidity", response["main"]["humidity"])
        citypy_df.set_value(index, "Lat", response["coord"]["lat"])
        citypy_df.set_value(index, "Lng", response["coord"]["lon"])
        citypy_df.set_value(index, "Max Temp", response["main"]["temp_max"])
        citypy_df.set_value(index, "Wind Speed", response["wind"]["speed"])
        
        
        print("Processing Record City # " , row_count,  ' | ' , response["name"])
        row_count += 1
        
print("------------------------")
print("Data Retrieval Complete")
print("------------------------")
        
```

```python
#Latitude vs. Temperature Plot
plt.scatter(citypy_df["Lat"], citypy_df["Max Temp"], edgecolor="black", linewidths=1, marker="o", cmap='GnBu', alpha=1)

plt.title("City Latitude vs. Max Temperature (08/22/18)")
plt.ylabel("Max Temperature(F)")
plt.xlabel("Latitude")
plt.grid()
plt.xlim([-60, 85])
plt.ylim([-60, 120])

# Save the figure
plt.savefig("Output/Lat_Temp.png")

# Show plot
plt.show()
```
